# Data Preprocessing Tools

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

## Importing the dataset

```python
# Import the dataset from a path in your project directory and store it to a variable
dataset = pd.read_csv('Data/Data.csv')

# Get all the rows from the dataset and all columns except the dependent variable vector (last). These may need to be updated
X = dataset.iloc[:, :-1].values 

# Get all rows and values from the LAST column (where the dependent variable lies)
y = dataset.iloc[:, -1].values 
```

```python
# Check values of X for correctness
print(X)
```

```python
# Check values of y for correctness
print(y)
```

## Taking care of missing data

```python
# Import imputer
from sklearn.impute import SimpleImputer 

# Replace missing values with mean of all values in column
imputer = SimpleImputer(missing_values=np.nan, strategy='mean') 

# Fit the data
imputer.fit(X[:, 1:3]) 

# Update the missing entries
X[:, 1:3] = imputer.transform(X[:, 1:3]) 
```

```python
# Check if the missing data has been replaced
print(X)
```

## Encoding categorical data

### Encoding the Independent Variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
# Check the one hot encoding of categorical data
print(X)
```

### Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Check dependent variable vector
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
# Import scikitlearn library
from sklearn.model_selection import train_test_split

# Split the data into training and test sets. Modify parameters as needed according to scikitlearn documentation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Check the values of training set for independent variables
print(X_train)
```

```python
# Check the values of test set for independent variables
print(X_test)
```

```python
# Check the values of training set for dependent variable
print(y_train)
```

```python
# Check the values of test set for dependent variable
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```
